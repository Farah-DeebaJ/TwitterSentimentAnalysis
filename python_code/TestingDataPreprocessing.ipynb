```python
# no of rows in the dataset
rows = test_df.count()
print('no of rows in the test data set : ', rows[0])
```

```python
emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    "]+", flags=re.UNICODE)
```

```python
SentimentText = []
def preprocess(df):
    # i = 0
    for i in range(0,rows[0]):
        text = df['SentimentText'][i]
        text = re.sub(r'@[A-Za-z0-9]+','',text)
        text = re.sub("[^a-zA-Z]", " ", text)
        text =emoji_pattern.sub(r'', text)
        text =text.lower()
        text = re.sub(r'[^\w\s]','',text)
        text = re.sub(r'(.)\1+', r'\1\1', text)
        tokens = word_tokenize(text)
        new_text =""
        for token in tokens:
            new_text += PorterStemmer().stem(token)
            new_text +=' '
        SentimentText.append(new_text)

        if i%10000 == 0:
            print('processed :' , i , 'tweets')
        # i = i + 1
    print('Done PreProcessing the text' )
    return SentimentText
```

```python
curr_time = time.time()
preprocess(test_df)
new_time = time.time()

time_taken = (new_time - curr_time)/60
print('time taken : ',time_taken, ' mins')
```
